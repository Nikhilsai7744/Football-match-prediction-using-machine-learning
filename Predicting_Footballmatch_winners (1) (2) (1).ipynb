## Cleaning data for machine learning

```python
matches.dtypes # checking datatype for each coloumn, ML algorithm can't work with object data
```

```python
#converting the date coloumn string into datatype

matches["date"]=pd.to_datetime(matches["date"])
```

## Creating predictors for machine learning

```python
# we need venue coloumn as a numbers,so we convert string into category and then to numbers

matches["venue_code"]=matches["venue"].astype("category").cat.codes
```

```python
matches #here in venuecode '0' indicates away and '1' indicates home match
```

```python
#generating a unique code for the opponents same as we did before

matches["opp_code"]=matches["opponent"].astype("category").cat.codes
```

```python
#convert the hour coloumn to an integer using regular expression, so that we use number as input in ML 
matches["hour"]=matches["time"].str.replace(":.+","",regex="true").astype("int")
```

```python
# we create a day_code coloumn to generate a number for a day coloumn
matches["day_code"]=matches["date"].dt.dayofweek #using dayofweek property we get a number for particular day in week
```

```python
# creating a number for a result coloumn using true false expression
matches["target"]=(matches["result"]=="W").astype("int")
```

```python
#In hour and opp_code coloumn we can see the number
#In daycode we got no of day of each week
matches 
```

## Creating initial machine learning model

```python
#it's ready to train our ml model
#importing random forestclassifier model they can pickup non-linearities in data
from sklearn.ensemble import RandomForestClassifier
```

```python
#initializing rfc class by passing parameters
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1) #set a no of decision tree value
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"]) #.fit method trains rf models using predictors
```

```python
preds = rf.predict(test[predictors]) #generating prediction by passing test data
```

```python
#import accuracy_score librabry to get accurate results
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

# Improving precision with rolling averages 

```python
grouped_matches = matches.groupby("team") #creates a one df for every squad in data
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date") #it gives a single group from data
```

```python
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group #this fun will take a set of coloumns to compute rolling averages for and then it'll take in a set of new coloumns that we want to assign the rolling averages 
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

new_cols #cloumns list to compute rolling averages for
```

```python
rolling_averages(group, cols, new_cols) #for the new coloumns that have info about what happened inthe last 3 matches
```

```python
#apply this to all the our matches
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols)) #applied func to each of those team df
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team') #this drops the extra index level from above
```

```python
matches_rolling
```

```python
#there are some repeated values in index, we want unique values from index
matches_rolling.index = range(matches_rolling.shape[0]) #thus will assign values from 0 to 1316
```

# Retraining ML model

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", 
              "Manchester United": "Manchester Utd", 
              "Newcastle United": "Newcastle Utd", 
              "Tottenham Hotspur": "Tottenham", 
              "West Ham United": "West Ham", 
              "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```
